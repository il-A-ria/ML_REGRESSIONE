### ESPLORAZIONE DELLA VARIABILE TARGET

```python
### usare sns.displot() per plottare la distribuzione (ricordate di passare un numero adeguato di bins)
import seaborn as sns
sns.displot(data=df['selling_price'], bins = 50);
```

```python
### fare un print di skweness e kurtosis
print('Skewness:', df['selling_price'].skew())
print('Kurtosis:', df['selling_price'].kurt())
```

### RELAZIONI CON ALTRE VARIABILI NUMERICHE

```python
### fare lo scatterplot() di ogni variabile numeriche (km_driven, seats, max_power, engine, mileage) contro il target
#selling_price VS max_power
for col in ['km_driven', 'seats', 'max_power', 'engine', 'mileage']:
  sns.scatterplot(x=col, y='selling_price', data=df);
  plt.show()
```

### RELAZIONI CON ALTRE VARIABILI CATEGORICHE

```python
### usare sns.boxplot per plottare ogni variabile categorica (year, fuel, seller_type, transmission, owner) contro il target

for col in ['year', 'fuel', 'seller_type', 'transmission', 'owner', 'torque']:
    
    sns.boxplot(x=col, y='selling_price', data=df)
    
    plt.xticks(rotation=90, fontsize=13) # ruota le etichette sull'asse x
    plt.show()
```

### CORRELAZIONI

```python
# plottare la matrice di correlazione con sns.heatmap() e individuare le variabili più correlate con selling_price, 
#oltre che quelle che sembrano essere collineari tra loro


corr_matrix = df.corr()    #matrice di correlazione

plt.figure(figsize=(10,9))
sns.heatmap(corr_matrix)     
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Volendo Seaborn mette a disposizione la funzione pairplot() per visualizzare gli scatterplot incrociati e le distribuzioni delle variabili...

```python
### plottare il pairplot di seaborn
sell_corr = corr_matrix['selling_price'].abs().sort_values(ascending=False)
sell_corr.iloc[1:11]
most_corr_vars = sell_corr.abs().iloc[0:10].index
most_corr_matrix = df[most_corr_vars].corr()

sns.pairplot(data=df[most_corr_vars])
```
